```python
# Importing the needed packages

from qiskit import QuantumCircuit
import numpy as np
import matplotlib.pyplot as plt
import os
from qiskit.quantum_info import Statevector, partial_trace
import qiskit.qasm2
from qiskit_aer import AerSimulator, Aer
from qiskit import transpile
from qiskit.quantum_info import DensityMatrix
```

```python
# Comparing the QAOA

QAOAnostra = qiskit.qasm2.load('qaoa.qasm', include_path=('.',), include_input_directory='append', custom_instructions=(), custom_classical=(), strict=False)
QAOAbase = qiskit.qasm2.load('qaoa_base.qasm', include_path=('.',), include_input_directory='append', custom_instructions=(), custom_classical=(), strict=False)

#print(QAOAnostra.draw())
initial_state = Statevector.from_label('001011')
final_state1=initial_state.evolve(QAOAnostra)
reduced_state1 = partial_trace(final_state1, [4,5])
display(reduced_state1.draw("latex"))

#print(QAOAbase.draw())
initial_state = Statevector.from_label('1011')
final_state2=initial_state.evolve(QAOAbase)
statevector_data = final_state2.data
density_matrix = DensityMatrix( np.outer(statevector_data, np.conj(statevector_data)) ) 
display(density_matrix.draw("latex"))
```
